to read the dataset we decided to consider only the columns that we thought were most useful, the columns are: id, title, year, authors, n_citation, doc_type, reference_count, referencesm doi, publisher

```python
import ijson
import csv
import numpy as np
import time

# Check if it's a small dataset
small_dataset = False

if small_dataset == True:
    # Start measuring time
    start = time.process_time()

    # Lists to store paper and author information
    PAPER = []
    Author = []

    # Open the JSON file for reading and a CSV file for writing
    with open('/content/drive/MyDrive/Colab Notebooks/università/ADM/Homework_5/dataset/dblp.v12.json', "rb") as f, open("./output.csv", "w", newline="") as csvfile:
        # Define the CSV fieldnames
        fieldnames = ['id', 'title', 'year', 'authors', 'n_citation', 'doc_type',
                      'reference_count', 'references', 'doi', 'publisher']
        # Create a CSV DictWriter
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames, escapechar='\\')
        # Write the header to the CSV file
        writer.writeheader()

        # Iterate over each element in the JSON file
        for i, element in enumerate(ijson.items(f, "item")):
            # Dictionary to store paper information
            paper = {}

            # Extract relevant information from the JSON element
            paper['id'] = element['id']
            paper['title'] = element['title']

            # Handle missing 'year' information
            year = element.get('year')
            if year:
                paper['year'] = year
            else:
                paper['year'] = np.nan

            # Handle missing 'authors' information
            author = element.get('authors')
            if author:
                paper['authors'] = author
            else:
                paper['authors'] = np.nan

            # Handle missing 'n_citation' information
            n_citation = element.get('n_citation')
            if n_citation:
                paper['n_citation'] = n_citation
            else:
                paper['n_citation'] = np.nan

            # Handle missing 'doc_type' information
            doc_type = element.get('doc_type')
            if doc_type:
                paper['doc_type'] = doc_type
            else:
                paper['doc_type'] = np.nan

            # Handle missing 'references' information
            references = element.get('references')
            if references:
                paper['reference_count'] = len(references)
                paper['references'] = ';'.join(str(int(r)) for r in references)
            else:
                paper['references'] = np.nan
                paper['reference_count'] = np.nan

            # Handle missing 'doi' information
            doi = element.get('doi')
            if doi:
                paper['doi'] = f"https://doi.org/{doi}"
            else:
                paper['doi'] = np.nan

            # Handle missing 'publisher' information
            publisher = element.get('publisher')
            if publisher:
                paper['publisher'] = publisher
            else:
                paper['publisher'] = np.nan

            # Write the paper information to the CSV file
            writer.writerow(paper)

            # Print progress every 46000 iterations
            if i % 46000 == 0:
                print(f"{i}:{round((time.process_time() - start), 2)}s ", end="")
```
